### Exaple usage

```python
from pyVCT import VCT
```

Define layer for simulation, by default energy configuration is taken from config files inside package. But you can provide your own using path2params argument

```python
layer = VCT(sizeX=0.3, sizeY=0.3, ncx=6, ncy=6, seed=48, fibers=True, fibers_dist=0.03)
```

Draw layer before simulation

```python
layer.draw()
```

Sumulate and draw result

```python
layer.simulate(100)
```

```python
layer.draw()
```
